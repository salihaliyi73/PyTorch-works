**TENSOR MATH OPERATIONS**

```python
import torch
```

```python
import numpy as np
```

```python
#Create tensor
tensor_a=torch.tensor([1,2,3,4])
tensor_b=torch.tensor([5,6,7,8])
tensor_a
tensor_b
```

**Addition**

```python
tensor_a + tensor_b
```

**Addition longhand**

```python
torch.add(tensor_a,tensor_b)
```

**substruction**

```python
tensor_a-tensor_b
```

```python
torch.sub(tensor_b,tensor_a)
```

```python
#multiplication
tensor_a*tensor_b
```

```python
torch.mul(tensor_b,tensor_a)
```

**division**
Automatically convert it to float

```python
tensor_b/tensor_a
```
