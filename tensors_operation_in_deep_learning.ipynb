```python
import torch
```

```python
my_torch =torch.arange(10)
```

```python
my_torch
```

```python
# Reshape and View
my_torch =my_torch.reshape(2,5)
my_torch
```

```python
# reshape if we don't know the number of items in tensors using -1
my_torch2=torch.arange(15)
my_torch2
```

```python
my_torch2=my_torch2.reshape(3,-1)
my_torch2
```

```python
my_torch2=my_torch2.reshape(-1,5)
my_torch2
```

```python
# VIEW
mytorch3=torch.arange(10)
mytorch3
```

```python
mytorch4=mytorch3.view(2,5)
mytorch4
```

```python
mytorch4=mytorch3.view(2,-1)
mytorch4
```
